```python
wd = "/Users/yn_mbp/Desktop/wd_namri"
nrrd = "/Users/yn_mbp/Desktop/nrrd"
scripts = "/Users/yn_mbp/git/yn-scripts"

nrrd_filename = "Na_correct_v6.nrrd"
reference_dicom = "03"
```

### ①working directoryの設定と、研究IDのリストの作成

```python
import os

#working directoryへ移動
os.chdir(wd)

#研究IDのリストを作成する
dir = [d for d in os.listdir() if os.path.isdir(os.path.join(".", d))]
dir
```

### ②nrrdファイルをworking directory内の各フォルダに移す

```python
%run $scripts/copy_file_to_matching_dir.py "$nrrd" "$wd" --file_name "$nrrd_filename"
```

### ③nrrdファイルをDICOMファイルに変換する
nrrdファイルから画像データを、元となるDICOMデータからDICOM metadataを抽出し、新たなDICOM dataを作ります

```python
corrected = nrrd_filename[:-5]

#nrrdファイルをdicomファイルに変換する。dicomファイルのメタデータの参考にする親ディレクトリを指定する（03）
for i in dir:
    os.chdir(f"./{i}")
    %run $scripts/nrrd2dcm.py $nrrd_filename $reference_dicom $corrected
    os.chdir("..")
```
